```python
from sklearn.datasets import load_wine
import pandas as pd

# Load wine dataset from sklearn
wine = load_wine()

# Convert to DataFrame
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['target'] = wine.target

# Save to CSV
df.to_csv('wine.csv', index=False)

# Display first few rows
df.head()
```

```python
# Extract input features (all columns except target)
X = df.drop('target', axis=1)

# Extract output (target) column
y = df['target']

# Optional: Show shapes to verify
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

# Set test size to get exactly 20 test samples
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=20/178)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_test
```

```python
#model evalution
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
